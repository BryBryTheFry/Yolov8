## Set up environment

- Create new environment
    - `conda create -n yolov8 python=3.9`
    - Not tested for other versions of python <br>
      
- Activate environment
    - `conda activate yolov8` <br>
      
- Enter correct folder <br>
      
- Clone github repo
    - `git clone https://github.com/ultralytics/ultralytics.git` <br>
      
- Enter repo foldder
    - `cd ultralytics` <br>
      
- Install required libraries
    - `pip install -e .`  <br>
      
- Download weights
    - Used https://github.com/ultralytics/assets/releases/download/v8.1.0/yolov8s.pt, https://github.com/ultralytics/assets/releases/download/v8.1.0/yolov8s-seg.pt, and https://github.com/ultralytics/assets/releases/download/v8.1.0/yolov8s-seg.pt
    - Move to a new folder "weights" <br>
